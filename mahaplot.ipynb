```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib widget
```

```python
x = [ 1,2,3,4]
y = [2,4,6,8]
y2 = [1,4,9,16]
plt.plot(x,y, label = "linear", color ='red', linestyle= "--", marker="o", linewidth=2)
plt.plot(x,y2, label = "square", color ='blue', linestyle= ":", marker="s", linewidth=2)
plt.title(" car")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid("true")
plt.legend()
plt.show()
```

```python
names= ["a","b","c","d"]
values= [15, 25, 5,25]
plt.bar(names, values)
plt.show()
```

```python
names= ["a","b","c","d"]
values= [15, 25, 5,25]
plt.barh(names, values)
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data,bins=30)
plt.show()
```

```python
x = [1,2,3,4]
y = [3 ,4,5,6]
plt.scatter(x,y)
plt.show()
```

```python
labels =["python", "java", "C++"]
size = [50,30,20]
plt.pie(size, labels=labels, autopct="%1.1f%%")
plt.show()
```

```python
names = ["a","b","c"]
values = [10,20,30]

x = [1,2,3,4]
y = [2,4,6,8]

m = ['maha','noha','ahmed','mohamed','naglaa']
n = [20,30,40,50,60]

fig, ax = plt.subplots(1,3)
#plt.figure(figsize(8,6), dpi = 100)
ax[0].plot(x,y)
ax[1].bar(names, values)
ax[2].pie(n, labels=m , autopct='%1.1f%%')
plt.show()
```

```python
plt.style.use("ggplot")
plt.plot(x,y)
plt.savefig("plot.png")
plt.show()
```

```python
df = pd.DataFrame({
    "x" :x,
    "y" : y
})
df.plot(x="x" , y ="y")
plt.show()
```

```python
plt.style.use("ggplot")
plt.plot(x,y)
plt.yscale("log")
plt.show()
```
